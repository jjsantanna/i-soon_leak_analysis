## Search Engine Setup

Run the following code to setup the search engine.

````python

# Load the data the data is located in chat_cn2en/chat_cn2en/x.xlsx where x are numbers from 2 to 41
# Load the data
data = []
for i in range(1, 42):
    try:
        data.append((pd.read_excel(f'{pathFromWorking}/{i}.xlsx'), f'{i}.xlsx'))
    except FileNotFoundError:
        pass

def search(search_term):
    # search the data
    search_results = []
    for f, name in data:
        messages = f['message_en']
        for i, m in enumerate(messages):
            if search_term.lower() in str(m).lower() or search_term in str(f['Time'][i]):
                search_results.append((name, i, f['Time'][i]))

    # sort the search results by date
    search_results = sorted(search_results, key=lambda x: x[2])


    print(f"Found {len(search_results)} mentions of the search term '{search_term}'")
    for name, i, date in search_results:
        print(f'{name} - {i}')
        file = 0
        for f, n in data:
            if n == name:
                file = f
                break
            
        for j in range(max(0, i-sentenceRange), min(i+sentenceRange, len(file))):
            print(f'{j}, from: {file["From"][j]} to: {file["To"][j]} - {file["Time"][j]}: {file["message_en"][j]}')

        print("\n\n")
````

## Code to translate the image .md file to english

The following code translates the Chinese text on images to English. We did this because all the images used in the .md file were in chinese and we wanted to translate them to english so we can use them in our research. Then we can do a more complete analysis of the data set. You might have to run this code multiple times to get all the images translated. Since googletrans Timeout error occurs sometimes.

````python
from cnocr import CnOcr
from googletrans import Translator
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
import os


# get all .md files
folder = 'i-s00n/0'
output_folder = 'i-s00n/2'
files = os.listdir(folder)
files = [f for f in files if f.endswith('.md')]

# read the files and extract the pngs
pngs = {}
for f in files:
    with open(f'{folder}/{f}', 'r') as file:
        try:
            lines = file.readlines()
            for l in lines:
                # remove <img src= and > and <br>
                l = l.replace("<img src=", "").replace("<br>", "").replace(">", "").replace("'", " ").replace("  ", " ").replace("\n", "").strip()

                list_of_files = l.split(" ")
                pngs[f] = list_of_files
        except Exception as e:
            pngs[f] = []
            


# extract the text from the pngs
texts = {}
for f, l in pngs.items():
    print(f)
    for i, p in enumerate(l):
        if p.endswith('.png') and not os.path.exists(f'{output_folder}/{p}'):
            img_fp = f'{folder}/{p}'
            ocr = CnOcr()
            text = ocr.ocr(img_fp)
            translator = Translator()
            img = Image.open(img_fp)
            for dict in text:
                dict['text'] = translator.translate(dest='en', text=dict['text']).text
           
                # draw white rectangle over the text
                try:
                    tuple1 = tuple(dict['position'][0])
                    tuple2 = tuple(dict['position'][2])

                    # swap tuple if the first element is greater than the second
                    if tuple1[0] > tuple2[0] or tuple1[1] > tuple2[1]:
                        tuple1, tuple2 = tuple2, tuple1

                    draw = ImageDraw.Draw(img)
                    draw.rectangle([tuple1, tuple2], fill='white')
                    # replace the text
                    size = 16
                    font = ImageFont.truetype("arial.ttf", int(size))
                    current_size = draw.textlength(dict['text'], font=font)
                    while draw.textlength(dict['text'], font=font) < abs(tuple2[0] - tuple1[0]):
                        size += 1
                    
                        font = ImageFont.truetype("arial.ttf", int(size))
                        if draw.textlength(dict['text'], font=font) == current_size:
                            break

                        current_size = draw.textlength(dict['text'], font=font)

                    
                    size -= 1
                    font = ImageFont.truetype("arial.ttf", int(size))


                    draw.text(tuple1, dict['text'], fill='black', font=font)

                except Exception as e:
                    print(e)


            img.save(f'{output_folder}/{p}')       
````
